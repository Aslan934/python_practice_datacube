```python

# 17. Aşağıda verilmiş kortecdə (tuple) 2-ci elementi çıxarın və uzunluğunu hesablayın
kortec = ("qırmızı", "mavi", "yaşıl")

# Korteci siyahıya (list) çeviririk
siyahı = list(kortec)

# İkinci elementi (1-ci indeksdəki "mavi" sözünü) silirik
siyahı.pop(1)

# Yenidən kortecə çeviririk
kortec = tuple(siyahı)

print(kortec)
print(len(kortec))
```
