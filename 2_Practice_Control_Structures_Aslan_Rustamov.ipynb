2.**while D√∂vr√ľ il…ô Toplama**
- sum_upto_n(n) adlńĪ funksiya yazńĪn:
- 1-d…ôn n-…ô q…ôd…ôr (n daxil olmaqla) b√ľt√ľn …ôd…ôdl…ôri toplayńĪn (bunun √ľ√ß√ľn while d√∂vr√ľnd…ôn istifad…ô edin).
- ToplamńĪ qaytarsńĪn.
- n m…ônfi olarsa, 0 qaytarńĪn v…ô ya x…ôb…ôrdarlńĪq mesajńĪ √ßap edin (√∂z se√ßiminizdir).

```python
def sum_upto_n(n):
    # KODUNUZU BURADA YAZIN
    total=0
    i=1
    if n<0:
      return 'musbet eded daxil edin'
    else:
      while i<=n:
        total+=i
        i+=1
    return total

# Test
test_values = [5, 10, -3, 0]
for val in test_values:
    print(f"sum_upto_n({val}) = {sum_upto_n(val)}")
```

3. **FizzBuzz**
- fizz_buzz(n) adlńĪ bir funksiya yaradńĪn v…ô 1-d…ôn n-…ô q…ôd…ôr d√∂vr edin:
- ∆Źg…ôr …ôd…ôd 3-…ô b√∂l√ľn√ľrs…ô, "Fizz",
- ∆Źg…ôr …ôd…ôd 5-…ô b√∂l√ľn√ľrs…ô, "Buzz",
- ∆Źg…ôr …ôd…ôd h…ôm 3-…ô, h…ôm d…ô 5-…ô b√∂l√ľn√ľrs…ô, "FizzBuzz",
- ∆Źks halda, …ôd…ôdin √∂z√ľn√ľ √ßap edin.

```python
def fizz_buzz(n):
    # KODUNUZU BURADA YAZIN
    for i in range(1,n+1):
      if i%3==0 and i%5==0:
        print("FizzBuzz")
      elif i%5==0:
        print("Buzz")
      elif i%3==0:
        print("Fizz")
      else:
        print(i)


fizz_buzz(15)
```

**Saitl…ôrin SayńĪnńĪ Tapmaq**
- count_vowels(text) adlńĪ funksiya yazńĪn:
- GiriŇü kimi s…ôtir (string) q…ôbul etsin.
- Saitl…ôrin (a, e, i, o, u v…ô eyni zamanda b√∂y√ľk h…ôrfl…ôr) sayńĪnńĪ - hesablamaq √ľ√ß√ľn d√∂vrd…ôn istifad…ô etsin.


```python
def count_vowels(text):
    # KODUNUZU BURADA YAZIN
    count=0
    vowels='aeuioAEUIO'
    for char in text:
      if char in vowels:
        count+=1
    return count

# Test
test_strings = ["Hello", "Education", "Rhythm", ""]
for t in test_strings:
    print(f"'{t}' s√∂z√ľnd…ô saitl…ôrin sayńĪ: {count_vowels(t)}")
```

5.
**S…ôtiri T…ôrsin…ô √áevirm…ôk**
- reverse_string(text) adlńĪ funksiya yaradńĪn:
- GiriŇü kimi s…ôtir q…ôbul etsin.
- Veril…ôn s…ôtirin t…ôrsini qaytarsńĪn.
- D√∂vr v…ô ya text[::-1] kimi qńĪsa √ľsullardan istifad…ô ed…ô bil…ôrsiniz.

```python
def reverse_string(text):
    # KODUNUZU BURADA YAZIN
    return text[::-1]


# Test
test_words = ["Python", "Colab", "Function", "12345"]
for w in test_words:
    print(f"{w} s…ôtirinin t…ôrsi -> {reverse_string(w)}")
```
